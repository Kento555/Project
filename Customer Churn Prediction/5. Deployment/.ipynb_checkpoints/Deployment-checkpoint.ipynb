# Capstone Project - Customer Churn Prediction - Deployment

```python
# Import package

import joblib
import numpy as np
import pandas as pd
from Preprocess import preprocess
import os
```

```python
# load prediction model

model = joblib.load(r"./4. Prepocess & Model Developement/model.sav")
```

```python
# Import data for production

dy = pd.read_csv('../2. Data/Telco-Customer-Churn - Deployment (Batch).csv')
```

```python
# Display data

dy.head()
```

```python
# Data preprocessing

preprocess_df = preprocess(dy)
```

```python
# Make prediction

prediction = model.predict(preprocess_df)
```

```python
# Prediction in array

prediction
```

```python
# Reformat prediction results to presentatble format.

prediction_df = pd.DataFrame(prediction, columns=["Predictions"])
prediction_df = prediction_df.replace({1:'Yes, the customer will terminate the service.', 
                                                    0:'No, the customer is happy with Telco Services.'})
```

```python
# Display output

prediction_df.head()
```

```python
# Combine Input data and prediction results

frame = [dy, prediction_df]
results = pd.concat(frame, axis=1, join='inner')
```

```python
# Display output

results.head()
```

```python
# Saving results to csv format (file Name is 'Results')

results.to_csv('Results.csv')
```
